```python
import scanpy as sc
import pandas as pd
```

```python
# Load your processed h5ad file
adata = sc.read_h5ad("/home/jemima/firstyear_report/MSc_processed_vento_tormo_data.h5ad")
```

```python
# Export expression matrix (X)
# Convert to dense if it's sparse
if hasattr(adata.X, "toarray"):
    X_df = pd.DataFrame(adata.X.toarray(), index=adata.obs_names, columns=adata.var_names)
else:
    X_df = pd.DataFrame(adata.X, index=adata.obs_names, columns=adata.var_names)
```

```python
# save or write to CSV
X_df.to_csv("adata_X.csv") 

# Step 3: Export cell metadata (obs)
adata.obs.to_csv("adata_obs.csv")

# Step 4: Export gene metadata (var)
adata.var.to_csv("adata_var.csv")
```

```python
print("Data shape:", X_df.shape)  # Should match adata.shape
print(X_df.memory_usage(deep=True).sum() / 1e6, "MB")  # Approx size in MB
```
